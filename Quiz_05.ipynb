```python
angka = 1
batas = 50
total = 0

while angka <= batas: # While
    for i in range(angka, angka +10): # For
        if i % 3 == 0 and i <= batas:
          total += i
          print(i, end="  ")
    angka += 10

print("Jumlah seluruh bilangan kelipatan 3 dari 1 sampai 30 adalah:", total)
```
